```python
id_month = []
```

```python
for index, row in df.iterrows():
    if row['month'] == "January":
        id_month.append("1-January")
    elif row['month'] == "February":
        id_month.append("2-February")
    elif row['month'] == "March":
        id_month.append("3-March")
    elif row['month'] == "April":
        id_month.append("4-April")
    elif row['month'] == "May":
        id_month.append("5-May")
    elif row['month'] == "June":
        id_month.append("6-June")
```

```python
id_month
```

```python
df['id_month'] = id_month
```

```python
df.head()
```

```python
df.sort_values(by=['id_month'], inplace=True)
```
